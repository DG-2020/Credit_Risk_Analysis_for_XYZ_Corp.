```python
# Data Function to detect Missing Values
def missing_data(data_xyz):
    total = data_xyz.isnull().sum().sort_values(ascending=False)
    percent = (data_xyz.isnull().sum()/data_xyz.isnull().count()).sort_values(ascending=False)
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    return(missing_data.head(50))

missing_data(data_xyz)
```

## 5. Logistic Regression
It is a predictive analysis, which is used to describe data and to explain the relationship between one dependent binary variable and one or more nomnal, ordinal, interval or ratio-level independent variables. 

A binary logistic model has a dependent variable with two possible values, such as pass/fail which is represented by an indicator variable, where the two values are labeled "0" and "1".

In the logistic model, the log-odds (the logarithm of the odds) for the value labeled "1" is a linear combination of one or more independent variables ("predictors"); the independent variables can each be a binary variable (two classes [0 or 1], coded by an indicator variable) or a continuous variable (any real value).

The probability of the corresponding value will vary between '0' and '1'.
